```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv('./quality.csv')
train
```

```python
#split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(train.drop('PoorCare', axis=1), train['PoorCare'], test_size=0.15, random_state=88)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
X_train.head()
```

```python
X_train_new = pd.DataFrame(X_train, columns=['StartedOnCombination','ProviderCount'])
X_test_new = pd.DataFrame(X_test, columns=['StartedOnCombination','ProviderCount'])
```

```python
#What is the coefficient for "StartedOnCombination"?
model = LogisticRegression().fit(X_train_new, y_train)
model.coef_[0][0]
```

```python
model.score(X_train_new, y_train)
model.predict(X_test_new)
```

```python
model.score(X_test_new, y_test)
```
